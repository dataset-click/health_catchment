```python
import csv
import pandas as pd
import requests
import json

#check the distance from one address to all possibilities within 100km

df = pd.read_csv('knn_sample_ambo_y.csv')
df_loc=df[['address_detail_pid','addr_lat','addr_lon','id','hosp_lat','hosp_lon']]
```

```python
import time
time_milliseconds = 500
time_sec = 0.001 * time_milliseconds
filename='result_osm_ambo_y.csv'
def calc_dist(x_lat,x_lon,y_lat,y_lon):
    # call the OSMR API
    r = requests.get(f"http://router.project-osrm.org/route/v1/car/{x_lon},{x_lat};{y_lon},{y_lat}?overview=false""")
    # then you load the response using the json libray
    # by default you get only one alternative so you access 0-th element of the `routes`
    routes = json.loads(r.content)
    #print(type(routes))    
    #route_1 = routes.get("routes")[0]
    return routes
    #print(route_1['duration']/3600," hr"," | ",route_1['distance']/1000," km")
    
def get_txt(txt, route):
    try:
        route_1=route.get("routes")[0]
        dist_km=route_1['distance']/1000
        dist_hr=route_1['duration']/3600
        txt=txt+","+str(dist_km)+","+str(dist_hr)+"\n"
    except:
        txt=txt+"\n"
    return txt
    
        
    

file=open(filename,'w')
#print csv header
file.write('address_detail_pid,id,dist_km,dist_hr\n')
file.flush()
file.close()
i=0
#for index, row in df_loc.iloc[:2].iterrows():
for index, row in df_loc.iterrows():
#    print(row)
    file=open(filename,'a')
    txt=row['address_detail_pid']+","+str(row['id'])
    source_lat= row['addr_lat']
    source_lon= row['addr_lon']
    dest_lat= row['hosp_lat']
    dest_lon= row['hosp_lon']
    res=calc_dist(source_lat,source_lon,dest_lat,dest_lon)
    txt=get_txt(txt,res)
    i=i+1
    print(i,":",txt)    
    file.write(txt)
    file.flush()
    file.close()
#    time.sleep(time_sec)
```
