```python
import csv
import pandas as pd
import requests
import json
import googlemaps
from geopy.distance import geodesic

df = pd.read_csv('address_hospital_nt.csv')
#df = pd.read_csv('sample.csv')
df_loc=df[['address_detail_pid','addr_lat','addr_lon','nearest_blue','blue_lat','blue_lon','nearest_green','green_lat','green_lon','nearest_orange','orange_lat','orange_lon','nearest_red','red_lat','red_lon','nearest_ambo','ambo_lat','ambo_lon']]
gmaps = googlemaps.Client(key='XXXXX')  
```

```python
def calc_dist_gmaps(x_lat,x_lon,y_lat,y_lon):
    print(pd.isna(y_lat))
    if not pd.isna(y_lat):
        origin = (str(x_lat),str(x_lon))
        dest   = (str(y_lat),str(y_lon))
        distance = gmaps.distance_matrix([origin], [dest], mode='driving')['rows'][0]['elements'][0]
        print(origin, dest)
        #print(distance)
        hours = distance['duration']['value'] / 3600

        origin_gps = (x_lat, x_lon)
        destination_gps = (y_lat, y_lon)
        distance2 = geodesic(origin_gps, destination_gps).km
    else:
        hours=-1
        distance2=-1

    res = {
        "duration": hours,
        "distance": format(distance2)
        }

    return res
    #print(route_1['duration']/3600," hr"," | ",route_1['distance']/1000," km")

file=open('result_gmaps_nt.csv','w')
#print csv header
file.write('address_detail_pid,nearest_blue,blue_km,blue_hr,nearest_green,green_km,green_hr,nearest_orange,orange_km,orange_hr,nearest_red,red_km,red_hr,nearest_ambo,ambo_km,ambo_hr\n')

#for index, row in df_loc.iloc[:5].iterrows():
for index, row in df_loc.iterrows():
#    print(row)
    txt=row['address_detail_pid']+","
    source_lat= row['addr_lat']
    source_lon= row['addr_lon']
    dest_lat= row['blue_lat']
    dest_lon= row['blue_lon']
#    print("BLUE",source_lat,source_lon,dest_lat,dest_lon)
    res=calc_dist_gmaps(source_lat,source_lon,dest_lat,dest_lon)
    dist_km=res['distance']
    dist_hr=res['duration']
    #print("blue : ",dist_km," km ",dist_hr," hr")
    txt=txt+row['nearest_blue']+","+str(dist_km)+","+str(dist_hr)+","
    
    dest_lat= row['green_lat']
    dest_lon= row['green_lon']
#    print("GREEN",source_lat,source_lon,dest_lat,dest_lon)
    res=calc_dist_gmaps(source_lat,source_lon,dest_lat,dest_lon)
    dist_km=res['distance']
    dist_hr=res['duration']
    #print("green : ",dist_km," km ",dist_hr," hr")
    txt=txt+row['nearest_green']+","+str(dist_km)+","+str(dist_hr)+","
    
    dest_lat= row['orange_lat']
    dest_lon= row['orange_lon']
#    print("ORANGE",source_lat,source_lon,dest_lat,dest_lon)
    res=calc_dist_gmaps(source_lat,source_lon,dest_lat,dest_lon)
    dist_km=res['distance']
    dist_hr=res['duration']
    #print("orange : ",dist_km," km ",dist_hr," hr")
    txt=txt+row['nearest_orange']+","+str(dist_km)+","+str(dist_hr)+","
    
    dest_lat= row['red_lat']
    dest_lon= row['red_lon']
#    print("RED",source_lat,source_lon,dest_lat,dest_lon)
    res=calc_dist_gmaps(source_lat,source_lon,dest_lat,dest_lon)
    dist_km=res['distance']
    dist_hr=res['duration']
    #print("red : ",dist_km," km ",dist_hr," hr")
    txt=txt+str(row['nearest_red'])+","+str(dist_km)+","+str(dist_hr)+","
    
    dest_lat= row['ambo_lat']
    dest_lon= row['ambo_lon']
#    print("AMBO",source_lat,source_lon,dest_lat,dest_lon)
    res=calc_dist_gmaps(source_lat,source_lon,dest_lat,dest_lon)
    dist_km=res['distance']
    dist_hr=res['duration']
    #print("ambo : ",dist_km," km ",dist_hr," hr")
    txt=txt+row['nearest_ambo']+","+str(dist_km)+","+str(dist_hr)+"\n"
    
    print(txt)    
    file.write(txt)
file.close()    
```
